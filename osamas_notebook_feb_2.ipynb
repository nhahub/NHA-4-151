````python
import json

input_path = sft_data_path
output_path = "/content/drive/MyDrive/fine_tuning_cv_final/train_sharegpt_final.jsonl"

valid = 0
skipped = 0

with open(input_path, "r", encoding="utf-8") as infile, \
     open(output_path, "w", encoding="utf-8") as outfile:

    for line in infile:
        if not line.strip():
            continue

        try:
            rec = json.loads(line)
        except:
            skipped += 1
            continue


        if "messages" in rec:
            messages = rec["messages"]
        elif "conversations" in rec:
            messages = rec["conversations"]
        elif isinstance(rec, list):
            messages = rec
        else:
            skipped += 1
            continue

        new_conversations = []

        for msg in messages:
            role = msg.get("role") or msg.get("from")
            content = (msg.get("content") or msg.get("value") or "").strip()

            if not role or not content:
                continue


            if content.startswith("```"):
                content = content.replace("```json", "").replace("```", "").strip()


            if role in ["user", "human"]:
                role = "human"
            elif role in ["assistant", "gpt"]:
                role = "gpt"
            elif role == "system":
                role = "system"
            else:
                continue

            new_conversations.append({
                "from": role,
                "value": content
            })

        roles = {m["from"] for m in new_conversations}


        if "human" not in roles or "gpt" not in roles:
            skipped += 1
            continue

        outfile.write(
            json.dumps({"conversations": new_conversations}, ensure_ascii=False)
            + "\n"
        )
        valid += 1

print("Done ")
print("Valid samples:", valid)
print("Skipped:", skipped)
print("Saved to:", output_path)
````

````python
def generate_analysis_report(messages):
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    model_inputs = tokenizer([text], return_tensors="pt").to("cuda:0")

    generated_ids = model.generate(
        model_inputs.input_ids,
        max_new_tokens=1500,
        do_sample=False, top_k=None, temperature=None, top_p=None,
    )

    generated_ids = [
        output_ids[len(input_ids):]
        for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

    return response

response = generate_analysis_report(messages)
````
